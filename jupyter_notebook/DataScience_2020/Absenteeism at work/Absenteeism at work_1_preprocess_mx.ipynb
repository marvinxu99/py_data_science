```python
# Aftrer we are done with our check, we will drop the 'check' column
reason_columns = reason_columns.drop(['check'], axis=1)
reason_columns
```

# Grouping these variables = Classification

```python
# reason_columns.loc[:, 1:14]  or
reason_columns.loc[:, '1':'14']
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

### Extracting the Day of the Week from the 'Date' Column

```python
df_reason_mod['Date'][699].weekday()
```

```python
df_reason_mod['Date'][699]
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis=1)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod.columns
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[column_names_reordered]
df_reason_mod.head()
```

```python
df_reason_date_mod = df_reason_mod.copy()
```

```python
df_reason_date_mod['Education'].unique()   # 1 - high school, 2-undergraduate 3-graduate 4-post graduate
```
